```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/telco_churn_clean.csv")
```

```python
df["Churn"].value_counts()
```

```python
df["Churn"].value_counts().plot(kind="bar")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Number of Customers")
plt.title("Churn Distribution")
plt.show()
```

```python
sns.boxplot(x="Churn",y="tenure",data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.boxplot(x="Churn",y="MonthlyCharges",data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn"]).plot(kind="bar")
plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Customer Count")
plt.show()
```

```python
pd.crosstab(df["PaymentMethod"], df["Churn"]).plot(kind="bar")
plt.title("Payment Method vs Churn")
plt.show()
```

```python
pd.crosstab(df["InternetService"], df["Churn"]).plot(kind="bar")
plt.title("Internet Service vs Churn")
plt.show()
```

EDA Summary:

Low tenure customers have highest churn

Month-to-month contracts are risky

High monthly charges increase churn probability

Long-term contracts reduce churn

Payment method impacts churn
